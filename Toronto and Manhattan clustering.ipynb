# Analysis of both data together

In this notebook we analyze the data refering to Dowtown Toronto and Manhattan through the clustering methodology.

```python
import numpy as np 
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import json 
import requests
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
from pandas.io.json import json_normalize 
from scipy import *
from matplotlib import *
from pylab import *
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium # map rendering library
from bs4 import BeautifulSoup
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html
from yellowbrick.cluster import KElbowVisualizer
print('Libraries imported.')
```

# Load the dataset previously treated

```python
#retrived previously stored data
dtoronto_grouped = pd.read_csv("dtoronto_grouped.csv")
neighborhoods_venues_sorted = pd.read_csv("dtoronto_venues_sorted.csv")
toronto_data = pd.read_csv("dtoronto_data.csv")


manhattan_grouped = pd.read_csv("manhattan_grouped.csv")
neighborhoods_venues_sorted1 = pd.read_csv("manhattan_venues_sorted.csv")
manhattan_data = pd.read_csv("manhattan_data.csv")

dtoronto_grouped['City']='Toronto'
manhattan_grouped['City']='New York'
whole_grouped = dtoronto_grouped.append(manhattan_grouped,sort=False)
whole_grouped=whole_grouped.fillna(0)

#neighborhoods_venues_sorted['City']='Toronto'
#neighborhoods_venues_sorted1['City']='New York'
whole_neighborhoods_venues_sorted = neighborhoods_venues_sorted.append(neighborhoods_venues_sorted1,sort=False)
whole_neighborhoods_venues_sorted=whole_neighborhoods_venues_sorted.fillna(0)

toronto_data.drop(['Postal Code'],axis=1,inplace=True)

toronto_data['City']='Toronto'
manhattan_data['City']='New York'
whole_city_data = toronto_data.append(manhattan_data,sort=False)
whole_city_data=whole_city_data.fillna(0)

print(whole_grouped.shape)
print(dtoronto_grouped.shape)
print(manhattan_grouped.shape)
```

# Clustering: finding optimal k for kmeans methodology

In the next step we will use kmeans algorithm and obtain the optimal k using the elbow method.

```python
grouped_clustering = whole_grouped.drop(['Neighborhood','City'], 1)
```

```python
model = KMeans()
from yellowbrick.cluster import KElbowVisualizer
visualizer = KElbowVisualizer(model, k=(1,30))

visualizer.fit(grouped_clustering)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# set number of clusters
kclusters = 14

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(grouped_clustering)

# add clustering labels
try:
    whole_neighborhoods_venues_sorted = whole_neighborhoods_venues_sorted.drop('Cluster Labels',axis = 1)
except:
    None
whole_neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

whole_merged = whole_city_data

# merge dtoronto_grouped with dtoronto_data to add latitude/longitude for each neighborhood
whole_merged = whole_merged.merge(whole_neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

whole_merged.head() # check the last columns!
```

# Analyze individual Clusters

## Clusters
Display each cluster.

```python
for i in range(kclusters):
    print("Cluster "+str(i) + " with {} elements.".format(whole_merged.loc[whole_merged['Cluster Labels'] == i,
                                                                               whole_merged.columns[[1] + list(range(5, whole_merged.shape[1]))]].shape[0]))
    display(whole_merged.loc[whole_merged['Cluster Labels'] == i, whole_merged.columns[[1,4] + list(range(6,whole_merged.shape[1]))]])
```

# Mapping the clusters

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(whole_merged['Latitude'], whole_merged['Longitude'], whole_merged['Neighborhood'], whole_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
