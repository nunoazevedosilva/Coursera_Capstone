# Data of Dowtown Toronto Neighboorhoods

In this notebook we get the data refering to Dowtown Toronto, from coordinates to most common venues and generate tables and maps for the report. We conclude by saving the data in a file to be processed on another notebook. 

```python
import numpy as np 
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import json 
import requests
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
from pandas.io.json import json_normalize 
from scipy import *
from matplotlib import *
from pylab import *
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium # map rendering library
from bs4 import BeautifulSoup
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html
print('Libraries imported.')
```

## Get Toronto Data into a Pandas dataframe, cleaning it

```python
#scrapping wikipedia neighborhoods data
data_source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
data_soup=BeautifulSoup(data_source,'lxml')
from IPython.display import display_html
tab = str(data_soup.table)

df_0 = pd.read_html(tab)
df=df_0[0]

# Dropping Borough 'Not assigned'
df = df[df.Borough != 'Not assigned']

# Combining same Postalcode
df = df.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)

# Replacing neighbourhood 'Not assigned' with Borough
df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned',df['Borough'], df['Neighbourhood'])



#get location data
lat_data = pd.read_csv('https://cocl.us/Geospatial_data')

#merge with our data
df_new = pd.merge(df,lat_data,on='Postal Code')
df_new.rename(columns={'Neighbourhood':'Neighborhood'},inplace=True)
df_new.head()
```

## Clean the data, taking only the Toronto data and visualize them

```python
toronto_data = df_new[df_new['Borough'].str.contains('Toronto',regex=False)]

toronto_data.head()

print(toronto_data.shape)


# create map of downtown toronto using latitude and longitude values
map_toronto = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

## Define credentials and functions to retrieve venues from Foursquare API

```python
CLIENT_ID = 'K5Y421AZ0VQB5BVAYMRJGRL5PF44FA2ZGL4IAAKKKIEXVDMW' # your Foursquare ID
CLIENT_SECRET = 'PKCWCXYBX3FZXLZX0YMJMFCNNRXGOMD4IOTK0GTOER1NBVZQ' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
LIMIT = 100 # A default Foursquare API limit value


print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Get the data from Foursquare.

```python
dtoronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
```

## Analyze the data in terms of most common venues

```python
print("We retrieved the data for {} venues.".format(dtoronto_venues.shape[0]))
print('There are {} unique venue categories.'.format(len(dtoronto_venues['Venue Category'].unique())))

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
        
# one hot encoding
dtoronto_onehot = pd.get_dummies(dtoronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
dtoronto_onehot['Neighborhood'] = dtoronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [dtoronto_onehot.columns[-1]] + list(dtoronto_onehot.columns[:-1])
dtoronto_onehot = dtoronto_onehot[fixed_columns]

dtoronto_grouped = dtoronto_onehot.groupby('Neighborhood').mean().reset_index()

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = dtoronto_grouped['Neighborhood']

for ind in np.arange(dtoronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(dtoronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
#save the data to csv file
dtoronto_grouped.to_csv("dtoronto_grouped.csv", index=False)
neighborhoods_venues_sorted.to_csv("dtoronto_venues_sorted.csv", index=False)
toronto_data.to_csv("dtoronto_data.csv", index=False)
```

# Clustering: finding optimal k for kmeans methodology

In the next step we will use kmeans algorithm and obtain the optimal k using the elbow method.

```python
dtoronto_grouped_clustering = dtoronto_grouped.drop('Neighborhood', 1)
```

```python
model = KMeans()
from yellowbrick.cluster import KElbowVisualizer
visualizer = KElbowVisualizer(model, k=(1,12))

visualizer.fit(dtoronto_grouped_clustering)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# set number of clusters
kclusters = 4

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dtoronto_grouped_clustering)

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

dtoronto_merged = toronto_data

# merge dtoronto_grouped with dtoronto_data to add latitude/longitude for each neighborhood
dtoronto_merged = dtoronto_merged.merge(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

dtoronto_merged.head() # check the last columns!
```

# Mapping the clusters

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(dtoronto_merged['Latitude'], dtoronto_merged['Longitude'], dtoronto_merged['Neighborhood'], dtoronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
